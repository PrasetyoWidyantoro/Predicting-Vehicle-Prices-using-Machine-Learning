```python
# Prediksi
y_pred_train_dl = model.predict(X_train)
y_pred_test_dl = model.predict(X_test)
y_pred_valid_dl = model.predict(X_valid)

# MAE
mae_train_dl = mean_absolute_error(y_train, y_pred_train_dl)
mae_test_dl = mean_absolute_error(y_test, y_pred_test_dl)
mae_valid_dl = mean_absolute_error(y_valid, y_pred_valid_dl)

# R2
r2_train_dl = r2_score(y_train, y_pred_train_dl)
r2_test_dl = r2_score(y_test, y_pred_test_dl)
r2_valid_dl = r2_score(y_valid, y_pred_valid_dl)

# MSE
mse_train_dl = mean_squared_error(y_train, y_pred_train_dl)
mse_test_dl = mean_squared_error(y_test, y_pred_test_dl)
mse_valid_dl = mean_squared_error(y_valid, y_pred_valid_dl)

# RMSE
rmse_train_dl = np.sqrt(mse_train_dl)
rmse_test_dl = np.sqrt(mse_test_dl)
rmse_valid_dl = np.sqrt(mse_valid_dl)

# Melihat MAE, R2, MSE, dan RMSE pada data Train, Test, dan Validasi
print("MAE (Mean Absolute Error) Train: ", mae_train_dl)
print("MAE (Mean Absolute Error) Test: ", mae_test_dl)
print("MAE (Mean Absolute Error) Valid: ", mae_valid_dl)

print("R2 (R-squared) Train: ", r2_train_dl)
print("R2 (R-squared) Test: ", r2_test_dl)
print("R2 (R-squared) Valid: ", r2_valid_dl)

print("MSE (Mean Squared Error) Train: ", mse_train_dl)
print("MSE (Mean Squared Error) Test: ", mse_test_dl)
print("MSE (Mean Squared Error) Valid: ", mse_valid_dl)

print("RMSE (Root Mean Squared Error) Train: ", rmse_train_dl)
print("RMSE (Root Mean Squared Error) Test: ", rmse_test_dl)
print("RMSE (Root Mean Squared Error) Valid: ", rmse_valid_dl)
```
